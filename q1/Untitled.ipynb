```python
import json
```

```python
with open('data/auto_fill_data.json') as json_file:
    jsonData = json.load(json_file)
```

```python
jsonData
```

```python
import re
def clean_str(string):
    string = re.sub(r"[^A-Za-z0-9 ]", " ", string)
    string = re.sub(r"\'s", "", string)
    string = re.sub(r"\'ve", " \'ve", string)
    string = re.sub(r"n\'t", " n\'t", string)
    string = re.sub(r"\'re", " \'re", string)
    string = re.sub(r"\'d", " \'d", string)
    string = re.sub(r"\'ll", " \'ll", string)
    string = re.sub(r",", " , ", string)
    string = re.sub(r"\s{2,}", " ",string)
    return string.lower().strip()
```

```python
WORD_COUNT = {}
for data in jsonData:
    text = data['question']
    text = clean_str(text)
    for w in text.split():
        if w not in WORD_COUNT:
            WORD_COUNT[w] = 1
        else:
            WORD_COUNT[w] += 1
```

```python
def match(string,sentence):
    if re.search(string,sentence):
        return True
    else:
        return False        
```

```python
def getStrings(string):
    candidates = []
    for data in jsonData:
        if match(string, data['question']):
            candidates.append(data)
    print(candidates)
    sorted_list = sorted(candidates, key=lambda x: x['rank'], reverse=True)[:3]
```
